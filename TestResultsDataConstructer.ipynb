```python
kNums = [31]
allTheGuesses = []
for k in kNums:
    guesses = []
    print ("Doing testList")
    pCorrect = 0
    total = 0
    for y in range(testData.shape[0]):
        results = {}
        for z in range(trainData.shape[0]):
            distance = 0
            for i in range (4):
                distance = distance + pow(trainData.iloc[z, 7+i] - testData.iloc[y, 6+i], 2)
            distance = math.sqrt(distance)
            results[distance] = z
            if len(results) > k:
                sortedResults = sorted(results.keys())
                del results[sortedResults.pop()]
        monsterList = []
        for key in results.keys():
            monsterList.append(trainData.iloc[results.get(key), 6])
        sortedCount = Counter(monsterList).most_common(k)
        if k == 1:
            print ("For creature " + str(y) + " we think the creature is... " + sortedCount[0][0])
            guesses.append(sortedCount[0][0])
        else: 
            if len(Counter(monsterList).most_common(k)) == 1:
                print ("For creature " + str(y) + " we think the creature is... " + sortedCount[0][0])
                guesses.append(sortedCount[0][0])
            else: 
                while Counter(monsterList).most_common(k)[0][1] == Counter(monsterList).most_common(k)[1][1]:
                    monsterList.pop()
                    sortedCount = Counter(monsterList).most_common(k)
                    if len(Counter(monsterList).most_common(k)) == 1:
                        break
                print ("For creature " + str(y) + " we think the creature is... " + sortedCount[0][0])
                guesses.append(sortedCount[0][0])
    allTheGuesses.append(guesses)
```

```python
import pandas as pd
import csv
import math

from sklearn.datasets import load_boston
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
import numpy as np
```

```python
trainData = pd.read_csv('train.csv')
```

```python
trainData
```

```python
trainList = []
testList = []
kf = KFold(371, n_folds=10, shuffle=True, random_state = 11)
for train, test in kf:
    trainList.append(train)
    testList.append(test)
trainList
```

```python
testList
```

```python
for x in range(10):
    print ("Doing testList" + str(x))
    for y in testList[x]:
        maxD = 99999999.0
        index = -1
        for z in trainList[x]:
            distance = 0
            for i in range (4):
                distance = distance + pow(trainData.iloc[z, 1+i] - trainData.iloc[y, 1+i], 2)
            distance = math.sqrt(distance)
            if distance < maxD:
                maxD = distance
                index = z
        print (index)
        print (maxD)
```

```python
trainData.max(0)
# id                    897
# bone_length      0.817001
# rotting_flesh    0.932466
# hair_length             1
# has_soul         0.935721
max_bone_length = trainData["bone_length"].max()
max_rotting_flesh = trainData["rotting_flesh"].max()
max_hair_length = trainData["hair_length"].max()
max_has_soul = trainData["has_soul"].max()
```

```python
trainData["norm_bone"] = trainData["bone_length"] / max_bone_length
trainData["norm_rotting_flesh"] = trainData["rotting_flesh"] / max_rotting_flesh
trainData["norm_hair_length"] = trainData["hair_length"] / max_hair_length
trainData["norm_has_soul"] = trainData["has_soul"] / max_has_soul
```
